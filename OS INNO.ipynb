```python
space=[]
process=[]
v=int(input("No of Voides:"))
for x in range(v):
    print("Enter the Size of Viod/Space:",x+1)
    size_space=input()
    space.append(int(size_space))

p=int(input("No of Processes:"))
for x in range(p):
    print("Enter the Size of Process:",x+1)
    size_process=input()
    process.append(int(size_process))
    
print("The Sizes of Void are:",space)
print("The Sizes of Processes are:",process)
#for first fit
sf=space.copy()
sp=process.copy()
#for best fit
sb=space.copy()
bp=process.copy()
#for worst fit
sw=space.copy()
wp=process.copy()
#for next fit
sn=space.copy()
np=process.copy()
```

```python
#a void 
#b process
def first(a,b):
    index=[0]*len(b)
    for x in range(len(b)):
        flag=0
        #for assignment of process
        for y in range(len(a)):
            if a[y]>=b[x]:
                flag=1
                index[x]=y
                a[y]-=b[x]
                print("Process\t\tProcess Size\t\tAssigned Index")
                print("P",x+1,sep='',end='')
                print("\t\t",b[x],"\t\t\t",y+1)
                print("\nSize After Reduction:",end =" ")
                print(a)
                print("\n")
            if flag!=1:
                continue
            else:
                break
        else:
            print("Process\t\tProcess Size\t\tAssigned Index")
            print("P",x+1,sep='',end='')
            print("\t\t",b[x],"\t\t\t","Not Assigned")
            print("\nSize After No Reduction:",end =" ")
            print(a)
            print("\n")
            
    print("\t\tTABLE")
    print("Process\t\tProcess Size\t\tBlock Index")
    for x in range(len(b)):
        print("P",x+1,sep='',end='')
        print("\t\t",b[x],"\t\t\t",end =" ")
        if index[x]!=0:
            print(index[x]+1)
        else:
            print("Not Assigned.")    
    print("Minimun Hole:",min(a))
    print("Maximum Hole:",max(a))
    print("Total Space Present:",sum(a))
```

```python
def best(a,b):
    index=[0]*len(b)
    for x in range(len(b)):
        loc=-1
        for y in range(len(a)):
            if a[y]>=b[x]:
                if loc==-1:
                    loc=y 
                elif a[loc]>a[y]:
                    loc=y
            
        if loc!=-1:
            index[x]=loc
            a[loc]-=b[x]
            print("Process\t\tProcess Size\t\tAssigned Index")
            print("P",x+1,sep='',end='')
            print("\t\t",b[x],"\t\t\t",loc+1)
            print("\nSize After Reduction:",end =" ")
            print(a)
            print("\n")
        else:
            print("Process\t\tProcess Size\t\tAssigned Index")
            print("P",x+1,sep='',end='')
            print("\t\t",b[x],"\t\t\t","Not Assigned")
            print("\nSize After No Reduction:",end =" ")
            print(a)
            print("\n")
            
            
    print("\t\tTABLE")
    print("Process\t\tProcess Size\t\tBlock Index")
    for x in range(len(b)):
        print("P",x+1,sep='',end='')
        print("\t\t",bp[x],"\t\t\t",end =" ")
        if index[x]!=0:
            print(index[x]+1)
        else:
            print("Not Assigned.")
    print("Minimun Hole:",min(a))
    print("Maximum Hole:",max(a))
    print("Total Space Present:",sum(a))
   
```

```python
def worst(a,b):
    index=[-1]*len(b)
    for x in range(len(b)):
        maxs=-1
        for y in range(len(a)):
            if a[y]>=b[x]:
                if maxs==-1:
                    maxs=y
                elif a[maxs]<a[y]:
                    maxs=y
                    
        if maxs !=-1:
            index[x]=maxs
            a[maxs]-=b[x]
            print("Process\t\tProcess Size\t\tAssigned Index")
            print("P",x+1,sep='',end='')
            print("\t\t",b[x],"\t\t\t",maxs+1)
            print("\nSize After Reduction:",end =" ")
            print(a)
            print("\n")
        else:
            print("Process\t\tProcess Size\t\tAssigned Index")
            print("P",x+1,sep='',end='')
            print("\t\t",b[x],"\t\t\t","Not Assigned")
            print("\nSize After No Reduction:",end =" ")
            print(a)
            print("\n")
                
    print("\t\tTABLE")
    print("Process\t\tProcess Size\t\tBlock Index")
    for x in range(len(b)):
        print(x+1,"\t\t",wp[x],"\t\t\t",end =" ")
        if index[x]!=-1:
            print(index[x]+1)
        else:
            print("Not Assigned.")
    print("Minimun Hole:",min(a))
    print("Maximum Hole:",max(a))
    print("Total Space Present:",sum(a))
   
```

```python
def next(a,b):
    index=[-1]*len(b)
    flag=0
    c=0
    for x in range(len(b)):
        for y in range(len(a)):
            if y<c:
                continue
            else:
                if a[y]>=b[x]:
                    index[x]=y
                    flag=1
                    a[y]-=b[x]
                    c=y
                    break
                y=(y+1)%len(a)
                #CIRCULAR APPROACH
                for i in range(c,len(a)):
                    if a[i]>=b[x]:
                        flag=1
                        index[x]=i
                        a[i]-=b[x]
                        break
                    else:
                        flag=0
                        break
                for j in range(0,c):
                    if a[j]>=b[x]:
                        flag=1
                        index[x]=j
                        a[j]-=b[x]
                        break
                    else:
                        flag=0
                        break
                        
        if flag==1:
            print("Process\t\tProcess Size\t\tAssigned Index")
            print("P",x+1,sep='',end='')
            print("\t\t",b[x],"\t\t\t",c+1,)
            print("\nSize After Reduction:",end =" ")
            print(a)
            print("\n")
        
        if flag!=1:
            print("Process\t\tProcess Size\t\tAssigned Index")
            print("P",x+1,sep='',end='')
            print("\t\t",b[x],"\t\t\t","Not Assigned")
            print("\nSize After No Reduction:",end =" ")
            print(a)
            print("\n")
            
            
        
        
    print("\t\tTABLE")
    print("Process\t\tProcess Size\t\tBlock Index")
    for x in range(len(b)):
        print(x+1,"\t\t",np[x],"\t\t\t",end =" ")
        if index[x]!=-1:
            print(index[x]+1)
        else:
            print("Not Assigned.")
    print("Minimun Hole:",min(a))
    print("Maximum Hole:",max(a))
    print("Total Space Present:",sum(a))
                
```

```python
print("------------------------FIRST FIT------------------------")
first(sf,sp)
print("------------------------FIRST FIT------------------------")
```

```python
print("------------------------BEST FIT------------------------")
best(sb,bp) 
print("------------------------BEST FIT------------------------")
```

```python
print("------------------------WORST FIT------------------------")
worst(sw,wp) 
print("------------------------WORST FIT------------------------")
```

```python
print("------------------------NEXT FIT------------------------")
next(sn,np) 
print("------------------------NEXT FIT------------------------")
```
